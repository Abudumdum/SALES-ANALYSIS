```python
import pandas as pd 
import os
import glob
```

```python
csv_files = glob.glob('*.{}'.format('csv'))
csv_files
```

```python
dataframe1 = pd.concat([pd.read_csv(f) for f in  csv_files], ignore_index = True)
dataframe1.to_csv('dataframe_new1.csv', index = False)
dataframe1
```

```python
dataframe_new1 = pd.read_csv('dataframe_new1.csv')
dataframe_new1 
```

```python
dataframe_new1.sort_values(by='Order Date', ascending = True)
```

```python
dataframe_new1['Month'] = dataframe_new1['Order Date'].str[0:2]
dataframe_new1 = dataframe_new1.dropna(how = 'all')
#dataframe_new1.isna()
```

```python
dataframe_new1 = dataframe_new1.sort_values(by='Order Date', ascending = True)
```

```python
dataframe_new1 = dataframe_new1[dataframe_new1['Month'].str[0:2] != 'Or']
dataframe_new1
```

```python
dataframe_new1['Month'].astype('int32')
dataframe_new1.head(4)
```

```python
dataframe_new1['Quantity Ordered']= pd.to_numeric(dataframe_new1['Quantity Ordered'])
dataframe_new1['Price Each']= pd.to_numeric(dataframe_new1['Price Each'])

dataframe_new1['Sales'] = dataframe_new1['Quantity Ordered'] * dataframe_new1['Price Each']
dataframe_new1
```

```python
 results = dataframe_new1.groupby('Month').sum()
results
```

```python
import matplotlib.pyplot as plt 
months = range(1,13)
plt.bar(months, results['Sales'])

plt.xticks(months)
plt.ylabel='sales in USD'
plt.xlabel = 'Month number'
plt.show()
```
